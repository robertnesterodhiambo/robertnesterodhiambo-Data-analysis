```python
import tkinter as tk
import math

# Create the main application window
root = tk.Tk()
root.title("Scientific Calculator")
root.configure(bg="#f0f0f0")  # Set background color

# Function to update the display when a button is clicked
def update_display(value):
    current_text = display_var.get()
    # If the display shows "Error", clear it
    if current_text == "Error":
        current_text = ""
    # Append the clicked button value to the display
    display_var.set(current_text + value)

# Function to clear the display
def clear_display():
    display_var.set("")

# Function to evaluate the expression and display the result
def calculate():
    try:
        result = eval(display_var.get())
        display_var.set(str(result))
    except Exception as e:
        display_var.set("Error")

# Function to perform square root operation
def calculate_sqrt():
    try:
        result = math.sqrt(eval(display_var.get()))
        display_var.set(str(result))
    except Exception as e:
        display_var.set("Error")

# Function to perform trigonometric functions
def calculate_trig(func):
    try:
        radians = math.radians(eval(display_var.get()))
        if func == "sin":
            result = math.sin(radians)
        elif func == "cos":
            result = math.cos(radians)
        elif func == "tan":
            result = math.tan(radians)
        display_var.set(str(result))
    except Exception as e:
        display_var.set("Error")

# Function to perform exponential function
def calculate_exp():
    try:
        result = math.exp(eval(display_var.get()))
        display_var.set(str(result))
    except Exception as e:
        display_var.set("Error")

# Function to perform natural logarithm
def calculate_ln():
    try:
        result = math.log(eval(display_var.get()))
        display_var.set(str(result))
    except Exception as e:
        display_var.set("Error")

# Create a StringVar to store the display text
display_var = tk.StringVar()

# Create the display widget
display = tk.Entry(root, textvariable=display_var, justify="right", font=("Arial", 18), bg="#ffffff", bd=5)
display.grid(row=0, column=0, columnspan=5, sticky="ew", padx=10, pady=10)

# Create buttons for numbers, operators, and other functions
buttons = [
    ("7", 1, 0), ("8", 1, 1), ("9", 1, 2), ("/", 1, 3),
    ("4", 2, 0), ("5", 2, 1), ("6", 2, 2), ("*", 2, 3),
    ("1", 3, 0), ("2", 3, 1), ("3", 3, 2), ("-", 3, 3),
    ("0", 4, 0), (".", 4, 1), ("=", 4, 2), ("+", 4, 3),
    ("Clear", 5, 0), ("sin", 1, 4), ("cos", 2, 4), ("tan", 3, 4),
    ("exp", 4, 4), ("ln", 5, 4), ("sqrt", 5, 1)
]

# Create and place the buttons
for (text, row, col) in buttons:
    if text == "Clear":
        button = tk.Button(root, text=text, command=clear_display, width=5, height=2, bg="#ffcccc")
    elif text == "=":
        button = tk.Button(root, text=text, command=calculate, width=5, height=2, bg="#99ff99")
    elif text in ["sin", "cos", "tan"]:
        button = tk.Button(root, text=text, command=lambda t=text: calculate_trig(t), width=5, height=2, bg="#ccffcc")
    elif text == "sqrt":
        button = tk.Button(root, text=text, command=calculate_sqrt, width=5, height=2, bg="#ccffcc")
    elif text == "exp":
        button = tk.Button(root, text=text, command=calculate_exp, width=5, height=2, bg="#ccffcc")
    elif text == "ln":
        button = tk.Button(root, text=text, command=calculate_ln, width=5, height=2, bg="#ccffcc")
    else:
        button = tk.Button(root, text=text, command=lambda t=text: update_display(t), width=5, height=2, bg="#cccccc")
    button.grid(row=row, column=col, sticky="nsew", padx=5, pady=5)

# Configure row and column weights to make buttons expandable
for i in range(6):
    root.grid_rowconfigure(i, weight=1)
for i in range(5):
    root.grid_columnconfigure(i, weight=1)

# Run the main event loop
root.mainloop()
```
