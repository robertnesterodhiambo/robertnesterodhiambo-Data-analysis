```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm


# Independent variables
X1 = df[['Monthly Average Number of Posts Created']]  # First independent variable
X2 = df[['Monthly Average Number of Likes Given']]    # Second independent variable

# Dependent variable
y = df['Monthly Average Number of Views']  # Dependent variable

# Function to run linear regression
def run_linear_regression(X, y, variable_name):
    # Initialize the linear regression model
    model = LinearRegression()

    # Fit the model
    model.fit(X, y)

    # Make predictions
    y_pred = model.predict(X)

    # Calculate residuals
    residuals = y - y_pred

    # R value and Standard Error
    r2 = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    std_error = np.sqrt(mse)

    # Intercept (b0) and Slope (b1)
    intercept = model.intercept_
    slope = model.coef_[0]

    # Display results
    print(f'Linear Regression for {variable_name}:')
    print(f'Intercept (b0): {intercept}')
    print(f'Slope (b1): {slope}')
    print(f'R: {r2}')
    print(f'Standard Error: {std_error}\n')

    # Plot residuals
    plt.figure(figsize=(8, 6))
    plt.scatter(y_pred, residuals)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.title(f'Residual Plot for {variable_name}')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')
    plt.show()

# Running linear regression for the two independent variables
run_linear_regression(X1, y, 'Monthly Average Number of Posts Created')
run_linear_regression(X2, y, 'Monthly Average Number of Likes Given')

# PART B: Running linear regression by Gender
def run_regression_by_gender(X, y, variable_name):
    genders = df['Gender'].unique()

    for gender in genders:
        print(f'\n--- Regression for {variable_name} for Gender: {gender} ---')
        df_gender = df[df['Gender'] == gender]
        X_gender = df_gender[[X]]
        y_gender = df_gender[y]

        run_linear_regression(X_gender, y_gender, f'{variable_name} for {gender}')

# Run regression by gender if needed
run_regression_by_gender('Monthly Average Number of Posts Created', 'Monthly Average Number of Views', 'Posts Created')
run_regression_by_gender('Monthly Average Number of Likes Given', 'Monthly Average Number of Views', 'Likes Given')
```
