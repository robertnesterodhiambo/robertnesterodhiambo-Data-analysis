```python
import mysql.connector

# Establish a connection to the MySQL database
connection = mysql.connector.connect(
    host="falode3368-1.ch6kyu62mcnp.us-east-1.rds.amazonaws.com",
    user="Falode",
    password="1QAZ2wsx",
    database="cis3368DB"
)

# Function to execute SQL queries
def execute_query(query, data=None, commit=False):
    cursor = connection.cursor()
    if data:
        cursor.execute(query, data)
    else:
        cursor.execute(query)
    if commit:
        connection.commit()

# Define SQL queries to insert rows into each table
insert_queries = [
    "INSERT INTO Facility (name) VALUES (%s)",
    "INSERT INTO Classroom (capacity, name, facility_id) VALUES (%s, %s, %s)",
    "INSERT INTO Teacher (firstname, lastname, room, classroom_id) VALUES (%s, %s, %s, %s)",
    "INSERT INTO Child (firstname, lastname, age, room, classroom_id) VALUES (%s, %s, %s, %s, %s)"
]

# Data to insert into each table
facility_data = [("Facility 1",), ("Facility 2",)]
classroom_data = [(20, "Classroom 1", 1), (30, "Classroom 2", 2)]
teacher_data = [("John", "Doe", "Room 1", 1), ("Jane", "Smith", "Room 2", 2)]
child_data = [("Alice", "Doe", 5, "Room 1", 1), ("Bob", "Smith", 6, "Room 2", 2)]

# Execute SQL queries to insert rows into each table
for query, data in zip(insert_queries, [facility_data, classroom_data, teacher_data, child_data]):
    for row in data:
        execute_query(query, row, commit=True)
        print("Row inserted successfully.")

# Close the connection
connection.close()
```
